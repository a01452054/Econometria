```python
# Import required libraries
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.vector_ar.vecm import coint_johansen
import matplotlib.pyplot as plt
import yfinance as yf

# Set the date range
start_date = '2022-01-01'
end_date = '2025-10-10'

# Fetch stock data for PEP and KO
pep_df = yf.download('PEP', start=start_date, end=end_date)
ko_df = yf.download('KO', start=start_date, end=end_date)

# Extract closing prices
pep_close = pep_df['Close']
ko_close = ko_df['Close']

# Set proper datetime index with frequency
pep_close.index = pd.date_range(start=pep_close.index[0], periods=len(pep_close), freq='B')
ko_close.index = pd.date_range(start=ko_close.index[0], periods=len(ko_close), freq='B')

# Align both series by date index
combined_df = pd.concat([pep_close, ko_close], axis=1, join='inner')
combined_df.columns = ['PEP', 'KO']
combined_df.dropna(inplace=True)

# Function for unit root tests
def unit_root_tests(series, name):
    print(f"\nUnit Root Tests for {name}:")

    # ADF Test
    adf_result = adfuller(series)
    print("ADF Test:")
    print(f'ADF Statistic: {adf_result[0]:.4f}')
    print(f'p-value: {adf_result[1]:.4f}')
    print(f'Critical Values: {adf_result[4]}')

    # KPSS Test
    kpss_result = kpss(series)
    print("\nKPSS Test:")
    print(f'KPSS Statistic: {kpss_result[0]:.4f}')
    print(f'p-value: {kpss_result[1]:.4f}')
    print(f'Critical Values: {kpss_result[3]}')

# Perform unit root tests
unit_root_tests(combined_df['PEP'], "PEP")
unit_root_tests(combined_df['KO'], "KO")

# Cointegration test
def cointegration_test(df):
    result = coint_johansen(df, det_order=0, k_ar_diff=1)
    print("\nJohansen Cointegration Test:")
    print(f"Trace statistic: {result.lr1}")
    print(f"Critical values (90%, 95%, 99%): {result.cvt}")

    for i in range(len(result.lr1)):
        if result.lr1[i] > result.cvt[i, 1]:  # 95% critical value
            print(f"r = {i}: Cointegration exists at 95% confidence level")
        else:
            print(f"r = {i}: No cointegration at 95% confidence level")

# Run cointegration test
cointegration_test(combined_df)

# Function to find best ARIMA model
def find_best_arima(series, name, max_p=3, max_d=2, max_q=3):
    best_aic = float('inf')
    best_order = None

    for p in range(max_p + 1):
        for d in range(max_d + 1):
            for q in range(max_q + 1):
                try:
                    model = ARIMA(series, order=(p, d, q))
                    results = model.fit()
                    if results.aic < best_aic:
                        best_aic = results.aic
                        best_order = (p, d, q)
                except:
                    continue

    print(f"\nBest ARIMA model for {name}:")
    print(f"Order: {best_order}")
    print(f"AIC: {best_aic:.2f}")
    return best_order

# Find and fit best ARIMA models
pep_order = find_best_arima(combined_df['PEP'], "PEP")
ko_order = find_best_arima(combined_df['KO'], "KO")

# Fit final ARIMA models
pep_model = ARIMA(combined_df['PEP'], order=pep_order).fit()
ko_model = ARIMA(combined_df['KO'], order=ko_order).fit()

# Forecast next 30 periods
forecast_steps = 30
pep_forecast = pep_model.forecast(steps=forecast_steps)
ko_forecast = ko_model.forecast(steps=forecast_steps)

# Create forecast index using business day frequency
forecast_index = pd.date_range(start=combined_df.index[-1], periods=forecast_steps + 1, freq='B')[1:]

# Plot original series with forecasts
plt.figure(figsize=(12,6))
plt.plot(combined_df['PEP'], label='PEP Historical')
plt.plot(forecast_index, pep_forecast, label='PEP Forecast', color='red')
plt.plot(combined_df['KO'], label='KO Historical')
plt.plot(forecast_index, ko_forecast, label='KO Forecast', color='green')
plt.title('PEP and KO Closing Prices with Forecasts')
plt.legend()
plt.show()

# Plot detailed forecast with confidence intervals
def plot_forecast(model, series, name, steps=30):
    forecast_obj = model.get_forecast(steps=steps)
    forecast = forecast_obj.predicted_mean
    conf_int = forecast_obj.conf_int()

    forecast_index = pd.date_range(start=series.index[-1], periods=steps + 1, freq='B')[1:]

    plt.figure(figsize=(12,6))
    plt.plot(series, label=f'{name} Historical')
    plt.plot(forecast_index, forecast, label='Forecast', color='red')
    plt.fill_between(forecast_index,
                     conf_int.iloc[:, 0],
                     conf_int.iloc[:, 1],
                     color='pink',
                     alpha=0.3,
                     label='95% Confidence Interval')
    plt.title(f'{name} Price Forecast')
    plt.legend()
    plt.show()

# Generate detailed forecast plots
plot_forecast(pep_model, combined_df['PEP'], "PEP")
plot_forecast(ko_model, combined_df['KO'], "KO")

# Print forecast values
print("\nPEP Forecast Values (next 5 periods):")
print(pep_forecast[:5])
print("\nKO Forecast Values (next 5 periods):")
print(ko_forecast[:5])
```

Se realizaron pruebas de raíz unitaria para las series temporales de PEP y KO para evaluar su estacionalidad. utilizamos pruebas complementarias: la prueba ADF y la prueba KPSS.

Para el caso de PEP, la prueba ADF mostró un estadístico de -2.0786 con un valor p de 0.2532. Este resultado indica que no se puede rechazar la hipótesis nula de que la serie tiene una raíz unitaria, lo que sugiere que no es estacionaria. Por otra parte, la prueba KPSS mostró un estadístico de 1.3128 y un valor p de 0.01. Este valor supera los valores críticos establecidos, lo que lleva a rechazar la hipótesis nula de estacionalidad. Por lo tanto, ambass pruebas coinciden en que la serie de PEP no es estacionaria.

Para KO, los resultados fueron similares. La prueba ADF entregó un estadístico de -1.6646 con un valor p de 0.4495, lo que también impide rechazar la hipótesis nula de raíz unitaria. Esto nos indica que la serie no es estacionaria. La prueba KPSS, por otro lado, arrojó un estadístico de 3.3544 con un valor p de 0.01, lo que nuevamente supera todos los valores críticos y permite rechazar la hipótesis nula de estacionalidad.. En consecuencia, se concluye que la serie de KO tampoco es estacionaria.

Esto implica que sus propiedades estadísticas, como la media y la varianza, no se mantienen constantes en el tiempo.
```python
# Import required libraries
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.vector_ar.vecm import coint_johansen
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import yfinance as yf

# Set the date range
start_date = '2022-1-01'
end_date = '2025-09-30'

# Fetch stock data for PEP and KO
pep_df = yf.download('PEP', start=start_date, end=end_date)
ko_df = yf.download('KO', start=start_date, end=end_date)

# Extract closing prices
pep_close = pep_df['Close']
ko_close = ko_df['Close']

# Set proper datetime index with frequency
pep_close.index = pd.date_range(start=pep_close.index[0], periods=len(pep_close), freq='B')
ko_close.index = pd.date_range(start=ko_close.index[0], periods=len(ko_close), freq='B')

# Align both series by date index
combined_df = pd.concat([pep_close, ko_close], axis=1, join='inner')
combined_df.columns = ['PEP', 'KO']
combined_df.dropna(inplace=True)

# Function for unit root tests with interpretation
def unit_root_tests(series, name):
    print(f"\nUnit Root Tests for {name}:")

    adf_result = adfuller(series)
    print("ADF Test:")
    print(f'ADF Statistic: {adf_result[0]:.4f}')
    print(f'p-value: {adf_result[1]:.4f}')
    print(f'Critical Values: {adf_result[4]}')
    print("Interpretation:")
    if adf_result[1] < 0.05:
        print(f"  - p-value < 0.05: Reject null hypothesis - {name} is stationary")
    else:
        print(f"  - p-value >= 0.05: Fail to reject null - {name} may be non-stationary")

    kpss_result = kpss(series)
    print("\nKPSS Test:")
    print(f'KPSS Statistic: {kpss_result[0]:.4f}')
    print(f'p-value: {kpss_result[1]:.4f}')
    print(f'Critical Values: {kpss_result[3]}')
    print("Interpretation:")
    if kpss_result[1] < 0.05:
        print(f"  - p-value < 0.05: Reject null hypothesis - {name} is non-stationary")
    else:
        print(f"  - p-value >= 0.05: Fail to reject null - {name} may be stationary")

# Perform unit root tests
unit_root_tests(combined_df['PEP'], "PEP")
unit_root_tests(combined_df['KO'], "KO")

# Function to plot correlograms with interpretation
def plot_correlograms(series, name, lags=30):
    plt.figure(figsize=(12, 8))

    plt.subplot(2, 1, 1)
    plot_acf(series, lags=lags, ax=plt.gca())
    plt.title(f'ACF for {name}')

    plt.subplot(2, 1, 2)
    plot_pacf(series, lags=lags, ax=plt.gca())
    plt.title(f'PACF for {name}')

    plt.tight_layout()
    plt.show()

    print(f"\nCorrelogram Interpretation for {name}:")
    print("  - ACF: Total correlation at each lag (includes indirect effects)")
    print("  - PACF: Direct correlation at each lag (controls for earlier lags)")
    print("  - Significant spikes suggest strong autocorrelation")
    print("  - ACF decay suggests MA order; PACF cutoff suggests AR order")

# Plot correlograms for original series
plot_correlograms(combined_df['PEP'], "PEP Original")
plot_correlograms(combined_df['KO'], "KO Original")

# Difference the series
pep_diff = combined_df['PEP'].diff().dropna()
ko_diff = combined_df['KO'].diff().dropna()

# Plot correlograms for differenced series
plot_correlograms(pep_diff, "PEP Differenced")
plot_correlograms(ko_diff, "KO Differenced")

# Cointegration test with interpretation
def cointegration_test(df):
    result = coint_johansen(df, det_order=0, k_ar_diff=1)
    print("\nJohansen Cointegration Test:")
    print(f"Trace statistic: {result.lr1}")
    print(f"Critical values (90%, 95%, 99%): {result.cvt}")
    print("Interpretation:")
    for i in range(len(result.lr1)):
        if result.lr1[i] > result.cvt[i, 1]:
            print(f"  - r = {i}: Cointegration exists at 95% confidence level")
        else:
            print(f"  - r = {i}: No cointegration at 95% confidence level")
    if result.lr1[0] > result.cvt[0, 1]:
        print("Conclusion: PEP and KO are cointegrated - long-run equilibrium exists")
    else:
        print("Conclusion: No evidence of cointegration between PEP and KO")

# Run cointegration test
cointegration_test(combined_df)

# Function to find best ARIMA model with interpretation
def find_best_arima(series, name, max_p=3, max_d=2, max_q=3):
    best_aic = float('inf')
    best_order = None

    for p in range(max_p + 1):
        for d in range(max_d + 1):
            for q in range(max_q + 1):
                try:
                    model = ARIMA(series, order=(p, d, q))
                    results = model.fit()
                    if results.aic < best_aic:
                        best_aic = results.aic
                        best_order = (p, d, q)
                except:
                    continue

    print(f"\nBest ARIMA model for {name}:")
    print(f"Order: {best_order}")
    print(f"AIC: {best_aic:.2f}")
    print("Interpretation:")
    print(f"  - p={best_order[0]}: Autoregressive terms")
    print(f"  - d={best_order[1]}: Differencing for stationarity")
    print(f"  - q={best_order[2]}: Moving average terms")
    return best_order

# Find and fit best ARIMA models
pep_order = find_best_arima(combined_df['PEP'], "PEP")
ko_order = find_best_arima(combined_df['KO'], "KO")

pep_model = ARIMA(combined_df['PEP'], order=pep_order).fit()
ko_model = ARIMA(combined_df['KO'], order=ko_order).fit()

# Forecast next 30 periods
forecast_steps = 30
pep_forecast = pep_model.forecast(steps=forecast_steps)
ko_forecast = ko_model.forecast(steps=forecast_steps)

# Create forecast index
forecast_index = pd.date_range(start=combined_df.index[-1], periods=forecast_steps + 1, freq='B')[1:]

# Plot original series with forecasts
plt.figure(figsize=(12,6))
plt.plot(combined_df['PEP'], label='PEP Historical')
plt.plot(forecast_index, pep_forecast, label='PEP Forecast', color='red')
plt.plot(combined_df['KO'], label='KO Historical')
plt.plot(forecast_index, ko_forecast, label='KO Forecast', color='green')
plt.title('PEP and KO Closing Prices with Forecasts')
plt.legend()
plt.show()

# Detailed forecast plot with confidence intervals and interpretation
def plot_forecast(model, series, name, steps=30):
    forecast_obj = model.get_forecast(steps=steps)
    forecast = forecast_obj.predicted_mean
    conf_int = forecast_obj.conf_int()

    forecast_index = pd.date_range(start=series.index[-1], periods=steps + 1, freq='B')[1:]

    plt.figure(figsize=(12,6))
    plt.plot(series, label=f'{name} Historical')
    plt.plot(forecast_index, forecast, label='Forecast', color='red')
    plt.fill_between(forecast_index,
                     conf_int.iloc[:, 0],
                     conf_int.iloc[:, 1],
                     color='pink',
                     alpha=0.3,
                     label='95% Confidence Interval')
    plt.title(f'{name} Price Forecast')
    plt.legend()
    plt.show()

    # Forecast interpretation
    last_value = series.iloc[-1]
    mean_forecast = forecast.mean()
    print(f"\nForecast Interpretation for {name}:")
    print(f"Last observed value: {last_value:.2f}")
    print(f"Average forecast value: {mean_forecast:.2f}")
    print(f"Forecast change: {mean_forecast - last_value:.2f}")
    if mean_forecast > last_value:
        print("Trend: Upward forecast trend")
    elif mean_forecast < last_value:
        print("Trend: Downward forecast trend")
    else:
        print("Trend: Flat forecast trend")
    print(f"95% CI range at period {steps}: [{conf_int.iloc[-1, 0]:.2f}, {conf_int.iloc[-1, 1]:.2f}]")

# Generate detailed forecast plots and interpretations
plot_forecast(pep_model, combined_df['PEP'], "PEP")
plot_forecast(ko_model, combined_df['KO'], "KO")

# Plot correlograms for model residuals
plot_correlograms(pep_model.resid, "PEP ARIMA Residuals")
plot_correlograms(ko_model.resid, "KO ARIMA Residuals")

# Print forecast values
print("\nPEP Forecast Values (next 5 periods):")
print(pep_forecast[:5])
print("\nKO Forecast Values (next 5 periods):")
print(ko_forecast[:5])
```

```python
# Import required libraries
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller, zivot_andrews
import matplotlib.pyplot as plt
import yfinance as yf

# Set the date range
start_date = '2022-01-01'
end_date = '2025-10-10'

# Fetch stock data for PEP and KO
pep_df = yf.download('PEP', start=start_date, end=end_date)
ko_df = yf.download('KO', start=start_date, end=end_date)

# Extract closing prices
pep_close = pep_df['Close']
ko_close = ko_df['Close']

# Ensure proper datetime index with frequency

pep_close = pep_close.asfreq('B', method='pad')
ko_close = ko_close.asfreq('B', method='pad')

# Function for Phillips-Perron test with interpretation
def phillips_perron_test(series, name):
    print(f"\nPhillips-Perron Test for {name}:")
    pp_result = adfuller(series, regression='c', autolag='AIC')
    print(f'PP Statistic: {pp_result[0]:.4f}')
    print(f'p-value: {pp_result[1]:.4f}')
    print(f'Critical Values: {pp_result[4]}')
    print("Interpretation:")
    if pp_result[1] < 0.05:
        print(f"  - p-value < 0.05: Reject null hypothesis - {name} is stationary")
    else:
        print(f"  - p-value >= 0.05: Fail to reject null - {name} may be non-stationary")
    print("  - Note: Using ADF with constant and automatic lag selection to approximate PP test")
    print("  - PP test adjusts for serial correlation and heteroskedasticity non-parametrically")
    return pp_result

# Function for Zivot-Andrews structural break test with interpretation
def zivot_andrews_test(series, name):
    print(f"\nZivot-Andrews Structural Break Test for {name}:")
    za_result = zivot_andrews(series, regression='c', autolag='AIC')
    print(f'ZA Statistic: {za_result[0]:.4f}')
    print(f'p-value: {za_result[1]:.4f}')
    print(f'Critical Values: {za_result[2]}')
    print(f'Breakpoint Index: {za_result[3]}')
    print("Interpretation:")
    if za_result[1] < 0.05:
        print(f"  - p-value < 0.05: Reject null hypothesis - {name} has a structural break")
        print(f"  - Breakpoint at index {za_result[3]} (position in series)")
    else:
        print(f"  - p-value >= 0.05: Fail to reject null - No clear evidence of a structural break")
    print("  - ZA test allows for a single break in intercept and/or trend")
    return za_result

# Perform Phillips-Perron tests
pep_pp = phillips_perron_test(pep_close, "PEP")
ko_pp = phillips_perron_test(ko_close, "KO")

# Perform Zivot-Andrews tests
pep_za = zivot_andrews_test(pep_close, "PEP")
ko_za = zivot_andrews_test(ko_close, "KO")

# Plot series with breakpoints
def plot_series_with_breakpoint(series, name, breakpoint_idx):
    break_date = series.index[breakpoint_idx]

    plt.figure(figsize=(12, 6))
    plt.plot(series, label=f'{name} Closing Prices')
    plt.axvline(x=break_date, color='red', linestyle='--', label=f'Breakpoint ({break_date.date()})')
    plt.title(f'{name} Closing Prices with Structural Break')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

    print(f"\nPlot Interpretation for {name}:")
    print(f"  - Red dashed line indicates the detected structural break at index {breakpoint_idx}")
    print(f"  - Estimated break date: {break_date.date()}")

# Plot series with breakpoints using datetime index

plot_series_with_breakpoint(pep_close, "PEP", pep_za[3])
plot_series_with_breakpoint(ko_close, "KO", ko_za[3])

```
